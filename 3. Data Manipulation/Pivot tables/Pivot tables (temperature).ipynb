```python
import pandas as pd
import numpy as np

temperatures = pd.read_csv("temperatures.csv", index_col=0)
print(temperatures)
```

## Temperature change over time

You can access the components of a date (year, month and day) using code of the form dataframe["column"].dt.component. For example, the month component is dataframe["column"].dt.month, and the year component is dataframe["column"].dt.year.

```python
# Add a year column to temperatures
temperatures["date"] = pd.to_datetime(temperatures["date"])
temperatures["year"]= temperatures["date"].dt.year


# Pivot avg_temp_c by country and city vs year
temp_by_country_city_vs_year = temperatures.pivot_table("avg_temp_c", index=["country", "city"], columns="year", fill_value=0, margins=True).sort_index()

# See the result
print(temp_by_country_city_vs_year)
```

```python
# Subset for Egypt to India
temp_by_country_city_vs_year.loc["Egypt":"India"]
```

```python
# Subset for Egypt, Cairo to India, Delhi
temp_by_country_city_vs_year.loc[("Egypt", "Cairo"):("India", "Delhi")]
```

### Calculating on a pivot table

```python
# Get the worldwide mean temp by year
mean_temp_by_year = temp_by_country_city_vs_year.mean()
mean_temp_by_year
```

```python
# Filter for the year that had the highest mean temp
sorted_temp = mean_temp_by_year.sort_values(ascending=False)
print(sorted_temp.iloc[[0]])
#sorted_temp.info()
```

```python
# Filter for the year that had the highest mean temp - one-liner
print(mean_temp_by_year.sort_values(ascending=False).iloc[[0]])
```

```python
# Filter for the city that had the lowest mean temp
print(mean_temp_by_year.sort_values(ascending=False).iloc[[-1]])
```
