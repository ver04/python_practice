### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

avocado = pd.read_csv("avocado.csv")
avocado.head()
```

## Histogram

```python
avocado["avg_price"].hist()
plt.show()
```

```python
# we can specify amount of bins
avocado["avg_price"].hist(bins=15)
plt.show()
```

## Bar plots

### Avocado sales by type

```python
# Get the total number of avocados sold of each size
nb_sold_by_size = avocado.groupby("size")["nb_sold"].sum()
nb_sold_by_size
```

```python
nb_sold_by_size.plot(kind="bar")
plt.show()
```

### Avocado average sales by year

```python
avg_amount_sold = avocado.groupby("year")["nb_sold"].mean()
avg_amount_sold
```

```python
avg_amount_sold.plot(kind="bar", title="Average amount of avocados sold by year")
plt.show()
```

## Line graph

```python
small = avocado.loc[avocado["size"].isin(["small"])]
small
```

```python
small.plot(kind="line", x="date", y="avg_price")
plt.show()
```

```python
small.plot(kind="line", x="date", y="avg_price", rot="90")
plt.show()
```

Sales of all avocados by date

```python
# Get the total number of avocados sold on each date
nb_sold_by_date = avocado.groupby("date")["nb_sold"].sum()

# Create a line plot of the number of avocados sold by date
nb_sold_by_date.plot(kind="line", rot=90)
plt.show()
```

## Scatter plots

### Avocados supply and demand

```python
avocado.plot(kind="scatter", x="nb_sold", y="avg_price", title="Number of avocados sold vs. average price")
plt.show()
```

## Layering plots

### Small vs large avocados

```python
avocado[avocado["size"]=="small"]["nb_sold"].hist()
avocado[avocado["size"]=="large"]["nb_sold"].hist()
```

```python
avocado[avocado["size"]=="small"]["nb_sold"].hist(alpha=0.7)
avocado[avocado["size"]=="large"]["nb_sold"].hist(alpha=0.7)
plt.legend(["small", "large"])
plt.show()
```

### Conventional vs organic avocados

```python
# Histogram of conventional avg_price 
avocado[avocado["type"]=="conventional"]["avg_price"].hist(alpha=0.5, bins=20)

# Histogram of organic avg_price
avocado[avocado["type"]=="organic"]["avg_price"].hist(alpha=0.5, bins=20)

# Add a legend
plt.legend(["conventional", "organic"])

# Show the plot
plt.show()
```

We can see that on average, organic avocados are more expensive than conventional ones, but their price distributions have some overlap!
